# Boston dataset

## Loading data and training the model

Loading the boston house-prices dataset.

```python
boston = load_boston()
```

### Conversion to Pandas Data Frames

```python
X = pd.DataFrame(boston.data, columns = boston.feature_names)
y = pd.DataFrame(boston.target, columns = ['Price'])
```

```python
X.merge(y, left_index=True,right_index=True)
```

Preparing training and testing data by using train_test_split function. 70% for training and 30% for testing.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

## Scatter plot of the training data

```python
# plt.scatter(X_train, y_train, c='royalblue')
# plt.show()
```

## KNeighborsRegressor - Creating and training the model

```python
n = X_train.shape[0]

params = [{'n_neighbors': 3, 'weights': 'uniform'},
          {'n_neighbors': 3, 'weights': 'distance'},
          {'n_neighbors': np.sqrt(n).astype(int), 'weights': 'uniform'},
          {'n_neighbors': np.sqrt(n).astype(int), 'weights': 'distance'},
          {'n_neighbors': n, 'weights': 'uniform'},
          {'n_neighbors': n, 'weights': 'distance'}]
```

Explain weights

Explain np.sqrt(n).astype(int)

```python
for i, param in enumerate(params):
    model = KNeighborsRegressor(**params[i])
    model.fit(X_train, y_train)
    score = model.score(X_test, y_test)

    pca = PCA(n_components = 1)
    X_reduced = pca.fit_transform(X)
    X_train_reduced, X_test_reduced, y_train, y_test = train_test_split(X_reduced, y, test_size=0.3, random_state=0)
    plot_range = np.arange(min(X_train_reduced), max(X_train_reduced), 0.01)
    model.fit(X_train_reduced, y_train)

    pred = model.predict(plot_range.reshape(-1, 1))

    new_X, new_y = zip(*sorted(zip(plot_range.reshape(-1, 1), pred))) 
    plt.scatter(X_train_reduced, y_train, c='royalblue', label='Train data')
    plt.plot(new_X, new_y, c='g', label='Predictions')    
    plt.legend()
    plt.title("KNeighborsRegressor (n_neighbors = %i, weights = '%s')" % (params[i]['n_neighbors'], params[i]['weights']))
    plt.ylabel('Price')
    plt.xlabel('Number of rooms')

    plt.show()
    
    print('Score for weights=' + str(params[i]['weights']) + ' and n_neighbors=' + str(params[i]['n_neighbors']) + ': ' + str(round(score*100, 2)))

# model = xgboost.XGBRegressor(colsample_bytree=0.4,
#                  gamma=0,                 
#                  learning_rate=0.07,
#                  max_depth=3,
#                  min_child_weight=1.5,
#                  n_estimators=70,                                                                    
#                  reg_alpha=0.75,
#                  reg_lambda=0.45,
#                  subsample=0.6,
#                  seed=42) 

# model.fit(X_train.values.reshape(-1, 1), y_train)
```

We can see that the best score we get for n_neighbors=3 so we selecting for our final calculation

```python
n_neighbors = 3
weights = "uniform"

model = KNeighborsRegressor(weights=weights, n_neighbors=n_neighbors)
model.fit(X_train, y_train)
score = model.score(X_test, y_test)
print("Final model score is: ", round(score*100, 2), "%")
```
